```python
data=pd.read_excel(r"C:\Users\Yash\OneDrive\Desktop\Python_JN\Survey Monkey\Data - Survey Monkey initial excel cleaning.xlsx",sheet_name="Working_sheet")
data
```

#### Droping unwanted columns for simplicity of reading and transformaing data

```python
data_edit=data.copy()
data_edit.drop(columns=['Start Date - ',
 'End Date - ',
 'Email Address - ',
 'First Name - ',
 'Last Name - ',
 'Custom Data 1 - ',],inplace=True)
data_edit
```

#### Unpivoting data to have a row wise and more usuable format for further transformation

```python
id_variables = list(data_edit.columns)[:8]
value_variables = list(data_edit.columns)[8:]

data_unpivot = data_edit.melt(id_vars = id_variables , value_vars = value_variables,var_name = "Question+Subquestion",value_name="Answers")
data_unpivot
```

#### Now from here on we will be getting our desired columns out of given data to answer some requiered question.

```python
ques = pd.read_excel(r"C:\Users\Yash\OneDrive\Desktop\Python_JN\Survey Monkey\Data - Survey Monkey initial excel cleaning.xlsx",sheet_name="Question")
```

```python
ques.drop(columns = ['Questions','Subquestions'],inplace=True)

data_ques = data_unpivot.merge(ques,how="left",left_on = "Question+Subquestion",right_on="Question+Subquestion")
data_ques['Answers'].fillna("Not Answered",inplace = True)
data_ques
```

```python
ques_count_test = data_ques[data_ques['Answers'] != "Not Answered"]
ques_count_test
```

```python
responses = ques_count_test.groupby('Question')['Respondent ID - '].nunique().reset_index()
responses.rename(columns = {'Respondent ID - ':'Responses'},inplace=True)
```

```python
data_ques_2 = data_ques.merge(responses,how="left",left_on="Question",right_on='Question')
data_ques_2
```

```python
same_ans = data_ques_2.groupby(["Question+Subquestion","Answers"])["Respondent ID - "].nunique().reset_index()
same_ans.rename(columns = {"Respondent ID - ":"Same Answers"},inplace=True)
```

```python
data_ques_3 = data_ques_2.merge(same_ans,how="left",left_on=["Question+Subquestion","Answers"],right_on=['Question+Subquestion','Answers'])
data_ques_3
```

#### Renaming columns into more general format

```python
list(data_ques_3.columns)[:8]
```

```python
Final_output_2 = data_ques_3.rename(columns = {'Identify which division you work in. - Response':'Division Primary',
 'Identify which division you work in. - Other (please specify)':'Division Secondary',
 'Which of the following best describes your position level? - Response':"Position Level",
 'Which generation are you apart of? - Response':'Generation',
 'Please select the gender in which you identify. - Response':'Gender',
 'Which duration range best aligns with your tenure at your company? - Response':'Tenure Range',
 'Which of the following best describes your employment type? - Response':'Employment Type'})
```
